# Protocolo de Configuración de Entornos (Data Science)

Este Notebook sirve como guía y plantilla para configurar entornos virtuales estables, evitando conflictos de versiones entre `pandas` y `janitor`.

---

## Paso 1: Creación y Activación (Terminal)

Ejecuta estos comandos en tu terminal **Git Bash** dentro de la carpeta de tu nuevo proyecto:

````python
# bash
# 1. Crear entorno
python -m venv .venv

# 2. Activar entorno

source .venv/Scripts/activate
````

## Paso 2: Instalación del Combo Estable

Para evitar los errores de `pandas.core.col`, instala estas versiones específicas:

```bash
pip install "pandas==1.5.3" "pyjanitor==0.23.1" "xarray<2024.0.0" numpy matplotlib seaborn scikit-learn ipykernel
```

## Paso 3: Registro del Kernel para VS Code

Ejecuta esto para que el entorno aparezca en la lista desplegable de VS Code sin buscar rutas:

```bash
python -m ipykernel install --user --name=nuevo_proyecto --display-name "Python (Nuevo Proyecto)"
```

## Paso 4: Verificación de Librerías

Una vez seleccionado el Kernel arriba a la derecha, ejecuta esta celda para validar:

````python
import pandas as pd
import janitor
import numpy as np

print(f"Pandas: {pd.__version__}")
print("Todo en orden para empezar el proyecto.")
````

## Paso 5: Instalación de Librerías adicionales

Una vez seleccionado el Kernel arriba a la derecha

El flujo correcto cada vez que necesites un paquete nuevo (por ejemplo, plotly o requests), hazlo así:

1. Vuelve a la terminal (Git Bash).
2. Asegúrate de estar en la carpeta del proyecto y con el entorno activo: source .venv/Scripts/activate.
3. Instala el paquete ahí: pip install nombre-del-paquete.
4. Reinicia el Kernel en VS Code (el botón de la flecha circular).
5. Haz el import en tu celda de código.
