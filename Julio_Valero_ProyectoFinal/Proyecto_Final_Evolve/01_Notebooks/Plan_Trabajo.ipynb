# ESTRUCTURA PLAN DE TRABAJO DEL PROYECTO

## Bloque 1: Configuración e Infraestructura (Set Up)
* **Tarea 1.1: Ingesta y Auditoría Técnica**
    * Carga de fuentes de datos (conexión a BD o archivos planos).
    * Inspección de calidad inicial y validación de tipos de datos.

## Bloque 2: Protocolo de Calidad de Datos
* **Tarea 2.1: Limpieza Estructural**
    * Estandarización de nombres de columnas mediante `clean_names`.
    * Conversión de tipos y formateo de variables críticas.
* **Tarea 2.2: Gestión Condicional de Nulos**
    * Análisis de representatividad de valores ausentes.
    * Ejecución de descartes o imputaciones según importancia de negocio.
* **Tarea 2.3: Tratamiento de Valores Atípicos**
    * Detección de outliers y registros duplicados.

## Bloque 3: Construcción del Datamart Analítico
* **Tarea 3.1: Ingeniería de Datos**
    * Integración de tablas mediante cruces de información relacional.
* **Tarea 3.2: Creación de Variables de Negocio**
    * Cálculo de KPIs de rentabilidad y palancas de valor.

## Bloque 4: Análisis Exploratorio e Insights
* **Tarea 4.1: Generación de Insights**
    * Análisis de patrones y validación de hipótesis de negocio.
* **Tarea 4.2: Segmentación de Datos**
    * Identificación de focos de valor y zonas geográficas prioritarias.

## Bloque 5: Machine Learning Pipeline
* **Tarea 5.1: Preselección de Variables**
    * Evaluación de importancia mediante `mutual_info` y `RFE`.
* **Tarea 5.2: Tratamiento de Desbalanceo**
    * Análisis de la variable target y aplicación técnica de balanceo (`SMOTE`, `UnderSampling`) si se requiere.
* **Tarea 5.3: Modelización Multitarea**
    * Ejecución de modelos de Clasificación, Regresión y No Supervisado.
    * Optimización de hiperparámetros y métricas de rendimiento.

## Bloque 6: Industrialización y Despliegue
* **Tarea 6.1: Refactorización para Producción**
    * Transformación de código experimental a módulos productivos.
* **Tarea 6.2: Implementación de Reentrenamiento SOFT**
    * Creación de scripts para actualizar modelos manteniendo la lógica original.
* **Tarea 6.3: Desarrollo del Código de Ejecución**
    * Empaquetado final (`cloudpickle`) y despliegue del script de inferencia.

```python
import pandas as pd
import matplotlib
matplotlib.use("Agg")
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from matplotlib.patches import FancyBboxPatch
import numpy as np
from datetime import datetime, date

TODAY = datetime(2026, 3, 12)

tasks = [
    # WBS, Nombre, Tipo, Start, End, Nivel
    ("1",       "AIRBNB - PROYECTO FINAL",                                  "project",  "2026-01-01", "2026-05-09", 0),
    ("1.1",     "Bloque 1: Configuración e Infraestructura",                 "group",    "2026-01-01", "2026-01-02", 1),
    ("1.1.1.1", "Carga de fuentes de datos",                                 "task",     "2026-01-01", "2026-01-01", 2),
    ("1.1.1.2", "Inspección de calidad inicial",                             "task",     "2026-01-02", "2026-01-02", 2),
    ("1.2",     "Bloque 2: Protocolo de Calidad de Datos",                   "group",    "2026-01-05", "2026-02-10", 1),
    ("1.2.1.1", "Estandarización de nombres de columnas",                    "task",     "2026-01-05", "2026-01-06", 2),
    ("1.2.1.2", "Conversión de tipos y formateo",                            "task",     "2026-01-07", "2026-01-08", 2),
    ("1.2.3",   "Análisis de representatividad de nulos",                    "task",     "2026-01-09", "2026-01-16", 2),
    ("1.2.4",   "Descartes / imputaciones por importancia negocio",          "task",     "2026-01-19", "2026-01-30", 2),
    ("1.2.5.1", "Detección de outliers y duplicados",                        "task",     "2026-02-02", "2026-02-10", 2),
    ("1.3",     "Bloque 3: Construcción del Datamart Analítico",             "group",    "2026-02-11", "2026-02-27", 1),
    ("1.3.1.1", "Integración de tablas relacionales",                        "task",     "2026-02-11", "2026-02-17", 2),
    ("1.3.2.1", "Cálculo de KPIs de rentabilidad",                          "task",     "2026-02-18", "2026-02-27", 2),
    ("1.4",     "Bloque 4: Análisis Exploratorio e Insights",                "group",    "2026-03-02", "2026-03-20", 1),
    ("1.4.1.1", "Análisis de patrones y validación de hipótesis",            "task",     "2026-03-02", "2026-03-11", 2),
    ("1.4.2.1", "Identificación de focos de valor geográficos",              "task",     "2026-03-12", "2026-03-20", 2),
    ("1.5",     "Bloque 5: Machine Learning Pipeline",                       "group",    "2026-03-23", "2026-04-17", 1),
    ("1.5.1.1", "Preselección de variables (mutual_info / RFE)",             "task",     "2026-03-23", "2026-03-27", 2),
    ("1.5.2.1", "Tratamiento de desbalanceo (SMOTE / UnderSampling)",        "task",     "2026-03-30", "2026-04-03", 2),
    ("1.5.3.1", "Modelización: Clasificación, Regresión, No Supervisado",    "task",     "2026-04-06", "2026-04-14", 2),
    ("1.5.3.2", "Optimización de hiperparámetros y métricas",                "task",     "2026-04-14", "2026-04-17", 2),
    ("1.6",     "Bloque 6: Industrialización y Despliegue",                  "group",    "2026-04-20", "2026-05-09", 1),
    ("1.6.1.1", "Refactorización a módulos productivos",                     "task",     "2026-04-20", "2026-04-24", 2),
    ("1.6.2.1", "Scripts de reentrenamiento SOFT",                           "task",     "2026-04-27", "2026-05-01", 2),
    ("1.6.3.1", "Empaquetado (cloudpickle) y script de inferencia",          "task",     "2026-05-05", "2026-05-09", 2),
]

df = pd.DataFrame(tasks, columns=["wbs", "nombre", "tipo", "start", "end", "nivel"])
df["start"] = pd.to_datetime(df["start"])
df["end"]   = pd.to_datetime(df["end"]) + pd.Timedelta(days=1)

BLOQUE_COLORS = {
    "project": ("#1a1a2e", "#1a1a2e"),
    "1.1": ("#2196F3", "#BBDEFB"),
    "1.2": ("#9C27B0", "#E1BEE7"),
    "1.3": ("#00BCD4", "#B2EBF2"),
    "1.4": ("#FF9800", "#FFE0B2"),
    "1.5": ("#4CAF50", "#C8E6C9"),
    "1.6": ("#F44336", "#FFCDD2"),
}

def get_color(row):
    wbs = row["wbs"]
    bloque = wbs[:3] if len(wbs) >= 3 else wbs
    if row["tipo"] == "project":
        return BLOQUE_COLORS["project"][0], BLOQUE_COLORS["project"][0]
    colors = BLOQUE_COLORS.get(bloque, ("#607D8B", "#CFD8DC"))
    if row["tipo"] == "group":
        return colors[0], colors[0]
    return colors[1], colors[0]

def get_status(row):
    if pd.isna(row["start"]):
        return "sin_fecha"
    if row["end"] <= TODAY:
        return "completada"
    if row["start"] <= TODAY < row["end"]:
        return "en_curso"
    return "futura"

df["status"] = df.apply(get_status, axis=1)

fig, ax = plt.subplots(figsize=(20, 13))
fig.patch.set_facecolor("#F8F9FA")
ax.set_facecolor("#F8F9FA")

ypos = list(range(len(df)))

for i, row in df.iterrows():
    y = ypos[i]
    nivel = row["nivel"]
    tipo  = row["tipo"]
    bg_color, bar_color = get_color(row)
    status = row["status"]

    row_bg = "#FFFFFF" if i % 2 == 0 else "#F0F4F8"
    ax.barh(y, (df["end"].max() - df["start"].min()).days + 10,
            left=df["start"].min(), height=0.9,
            color=row_bg, alpha=0.4, zorder=0)

    if pd.notna(row["start"]):
        dur = (row["end"] - row["start"]).days
        height = 0.65 if tipo == "task" else 0.72

        if tipo == "project":
            ax.barh(y, dur, left=row["start"], height=0.3,
                    color="#1a1a2e", alpha=0.9, zorder=3)
        elif tipo == "group":
            ax.barh(y, dur, left=row["start"], height=height,
                    color=bar_color, alpha=0.85, zorder=3,
                    edgecolor="white", linewidth=0.5)
        else:
            alpha = 1.0 if status == "completada" else (0.9 if status == "en_curso" else 0.65)
            ax.barh(y, dur, left=row["start"], height=height,
                    color=bg_color, alpha=alpha, zorder=3,
                    edgecolor=bar_color, linewidth=1.2)

            if status == "completada":
                ax.barh(y, dur, left=row["start"], height=height,
                        color="none", zorder=4,
                        edgecolor=bar_color, linewidth=1.5,
                        hatch="////", alpha=0.3)
                mid = row["start"] + (row["end"] - row["start"]) / 2
                ax.text(mid, y, "✓ 100%", ha="center", va="center",
                        fontsize=6.5, color=bar_color, fontweight="bold", zorder=5)
            elif status == "en_curso":
                pct_done = (TODAY - row["start"]).days / dur
                ax.barh(y, dur * pct_done, left=row["start"], height=height,
                        color=bar_color, alpha=0.5, zorder=4)
                mid = row["start"] + (row["end"] - row["start"]) / 2
                ax.text(mid, y, f"~{int(pct_done*100)}%", ha="center", va="center",
                        fontsize=6.5, color="#333", fontweight="bold", zorder=5)

nombre = df["nombre"].tolist()
ax.set_yticks(ypos)
ax.set_yticklabels(nombre, fontsize=8.2)
ax.invert_yaxis()

for i, row in df.iterrows():
    y = ypos[i]
    nivel = row["nivel"]
    tipo  = row["tipo"]
    label = ax.get_yticklabels()[i] if i < len(ax.get_yticklabels()) else None
    tick = ax.yaxis.get_major_ticks()[i]
    if tipo == "project":
        tick.label1.set_fontweight("black")
        tick.label1.set_fontsize(10)
        tick.label1.set_color("#1a1a2e")
    elif tipo == "group":
        tick.label1.set_fontweight("bold")
        tick.label1.set_fontsize(8.5)
        tick.label1.set_color("#333")
    else:
        tick.label1.set_fontsize(7.8)
        tick.label1.set_color("#555")
        pad = "    " * nivel
        tick.label1.set_text(pad + tick.label1.get_text())

ax.axvline(x=TODAY, color="#E53935", linewidth=2, linestyle="--", zorder=6, alpha=0.9)
ax.text(TODAY, -0.7, f" HOY\n {TODAY.strftime('%d/%m/%Y')}",
        color="#E53935", fontsize=7.5, fontweight="bold", va="top", ha="left", zorder=7)

import matplotlib.dates as mdates
ax.xaxis.set_major_formatter(mdates.DateFormatter("%d %b"))
ax.xaxis.set_major_locator(mdates.WeekdayLocator(byweekday=0, interval=2))
plt.xticks(rotation=35, ha="right", fontsize=7.5)

ax.set_xlim(datetime(2026, 1, 1), datetime(2026, 5, 16))
ax.set_ylim(len(df) - 0.5, -1.2)

ax.grid(axis="x", linestyle=":", alpha=0.4, color="#888", zorder=1)
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
ax.spines["left"].set_visible(False)

legend_patches = [
    mpatches.Patch(color="#BBDEFB", label="Bloque 1: Configuración", edgecolor="#2196F3", linewidth=1.2),
    mpatches.Patch(color="#E1BEE7", label="Bloque 2: Calidad de Datos", edgecolor="#9C27B0", linewidth=1.2),
    mpatches.Patch(color="#B2EBF2", label="Bloque 3: Datamart", edgecolor="#00BCD4", linewidth=1.2),
    mpatches.Patch(color="#FFE0B2", label="Bloque 4: EDA e Insights", edgecolor="#FF9800", linewidth=1.2),
    mpatches.Patch(color="#C8E6C9", label="Bloque 5: ML Pipeline", edgecolor="#4CAF50", linewidth=1.2),
    mpatches.Patch(color="#FFCDD2", label="Bloque 6: Despliegue", edgecolor="#F44336", linewidth=1.2),
    mpatches.Patch(facecolor="#ddd", label="Completada (✓)", hatch="////", edgecolor="#999", linewidth=0.8),
    mpatches.Patch(color="#aaa", label="En curso (~%)", edgecolor="#555", linewidth=1),
    mpatches.Patch(color="#e8e8e8", label="Futura", edgecolor="#aaa", linewidth=0.8),
    mpatches.Patch(color="#E53935", label=f"Hoy: {TODAY.strftime('%d/%m/%Y')}"),
]
ax.legend(handles=legend_patches, loc="upper right", fontsize=7.5,
          framealpha=0.9, ncol=2, title="Leyenda", title_fontsize=8)

ax.set_title("Diagrama de Gantt — Proyecto Final Evolve: Análisis Inversión AirBnb Madrid",
             fontsize=13, fontweight="bold", color="#1a1a2e", pad=14)

fig.tight_layout()
output_path = r"C:\Users\Oscar\OneDrive - FM4\Escritorio\EVOLVE\Data Science\EVOLVE\Julio_Valero_ProyectoFinal\Proyecto_Final_Evolve\09_Otros\gantt_airbnb.png"
fig.savefig(output_path, dpi=180, bbox_inches="tight", facecolor=fig.get_facecolor())
plt.close(fig)
print(f"Gantt guardado en: {output_path}")
```
