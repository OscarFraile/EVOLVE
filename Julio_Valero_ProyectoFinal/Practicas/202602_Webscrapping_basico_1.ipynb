```python
# Importamos las bibliotecas necesarias para HTTP y parseo HTML
import requests
from bs4 import BeautifulSoup

# URL de la pagina a scrapear
url = "https://quotes.toscrape.com/"

# Hacemos la solicitud HTTP para obtener el HTML de la pagina
response = requests.get(url)

# Verificamos que la solicitud fue exitosa (HTTP 200)
if response.status_code == 200:
    print("Conexion exitosa.")

    # Creamos un objeto BeautifulSoup para analizar el HTML recibido
    soup = BeautifulSoup(response.content, "html.parser")

    # Buscamos todas las citas (etiquetas <span> con clase "text")
    # Buscamos todos los bloques de cita
    quotes = soup.select("div.quote")
    for quote in quotes:
        autor = quote.find("small", class_="author").get_text(strip=True)
        cita = quote.find("span", class_="text").get_text(strip=True)
        print("Autor:", autor, "; Cita:", cita)

else:
# Si la respuesta no fue 200, informamos el error
    print("Error al conectar con la pagina.")
```

```python
# Extraer citas, autores y etiquetas de cada bloque
import requests
from bs4 import BeautifulSoup

# URL de la página
url = "https://quotes.toscrape.com/"

# Realizamos la solicitud HTTP
response = requests.get(url)

if response.status_code == 200:
    soup = BeautifulSoup(response.content, "html.parser")
    
    # Encontramos todos los bloques de citas
    quotes = soup.find_all("div", class_="quote")

    # Iteramos sobre cada cita para extraer información
    for quote in quotes:
        text = quote.find("span", class_="text").text  # Texto de la cita
        author = quote.find("small", class_="author").text  # Autor
        tags = [tag.text for tag in quote.find_all("a", class_="tag")]  # Etiquetas
        
        # Imprimimos la información extraída
        print("Cita:", text)
        print("Autor:", author)
        print("Etiquetas:", ", ".join(tags))
        print("-" * 50)  # Separador
else:
    print("Error al conectar con la página.")
```
