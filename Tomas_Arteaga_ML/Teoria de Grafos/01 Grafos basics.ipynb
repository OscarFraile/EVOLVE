```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
import networkx as nx

G = nx.Graph()

V = {'Dublin', 'Paris', 'Milan', 'Rome'}
E = [('Milan','Dublin'), ('Milan','Paris'), ('Paris','Dublin'),('Milan','Rome')]

G.add_nodes_from(V)
G.add_edges_from(E)
```

```python
print(f"V = {G.nodes}")
print(f"E = {G.edges}")
```

```python
print(f"Graph Order: {G.number_of_nodes()}")
print(f"Graph Size: {G.number_of_edges()}")
print(f"Degree for nodes: { {v: G.degree(v) for v in G.nodes} }")
print(f"Neighbors for nodes: { {v: list(G.neighbors(v)) for v in G.nodes}}")
```

```python
ego_graph_milan = nx.ego_graph(G, "Milan")
print(f"Nodes: {ego_graph_milan.nodes}")
print(f"Edges: {ego_graph_milan.edges}")
```

```python
# Add new nodes and edges
new_nodes = {'London', 'Madrid'}
new_edges = [('London','Rome'), ('Madrid','Paris')]

G.add_nodes_from(new_nodes)
G.add_edges_from(new_edges)

print(f"V = {G.nodes}")
print(f"E = {G.edges}")
```

```python
plt.figure(figsize=(6, 4))
pos = nx.spring_layout(G, seed=42)  # Posiciones de los nodos

nx.draw(
    G, pos,
    with_labels=True,
    node_size=2000,
    node_color='lightblue',
    arrows=True,
    arrowstyle='->',
    arrowsize=20,
    font_size=12
)

plt.title("Digraph de Ciudades")
plt.show()
```

#### 2.2 Multigrafo

```python
directed_multi_graph = nx.MultiDiGraph()
undirected_multi_graph = nx.MultiGraph()

V = {'Dublin', 'Paris', 'Milan', 'Rome'}
E = [('Milan','Dublin'), ('Milan','Dublin'), ('Paris','Milan'), 
('Paris','Dublin'), ('Milan','Rome'), ('Milan','Rome')]

directed_multi_graph.add_nodes_from(V)
undirected_multi_graph.add_nodes_from(V)

directed_multi_graph.add_edges_from(E)
undirected_multi_graph.add_edges_from(E)
```

```python

plt.figure(figsize=(7,5))
pos = nx.spring_layout(directed_multi_graph, seed=42)

nx.draw_networkx_nodes(directed_multi_graph, pos, node_size=2000, node_color="lightgreen")
nx.draw_networkx_labels(directed_multi_graph, pos)

# Dibujar aristas múltiples con curvas
nx.draw_networkx_edges(
    directed_multi_graph, pos,
    arrows=True,
    arrowstyle='->',
    arrowsize=20,
    connectionstyle='arc3,rad=0.2',
    min_source_margin=20, 
    min_target_margin=20
)

plt.title("Directed MultiGraph")
plt.axis("off")
plt.show()
```

```python

plt.figure(figsize=(7,5))
pos = nx.spring_layout(undirected_multi_graph, seed=42)

nx.draw_networkx_nodes(undirected_multi_graph, pos, node_size=2000, node_color="lightgreen")
nx.draw_networkx_labels(undirected_multi_graph, pos)

# Aristas múltiples con curvas
nx.draw_networkx_edges(
    undirected_multi_graph, pos,
    connectionstyle='arc3,rad=0.2'
)

plt.title("Undirected MultiGraph")
plt.axis("off")
plt.show()
```

#### 2.3 Weighted graphs

#### Ponderados por el enlace

```python
G = nx.DiGraph()

V = {'Dublin', 'Paris', 'Milan', 'Rome'}
E = [('Milan','Dublin', 19), ('Paris','Milan', 8), ('Paris','Dublin', 11), ('Milan','Rome', 5)]

G.add_nodes_from(V)
G.add_weighted_edges_from(E)
```

```python

import networkx as nx
import matplotlib.pyplot as plt

plt.figure(figsize=(7,5))
#pos = nx.spring_layout(G, seed=42)
pos = nx.shell_layout(G)

# Nodos
nx.draw_networkx_nodes(G, pos, node_size=2000, node_color="lightgreen")
nx.draw_networkx_labels(G, pos, font_size=12)

# Aristas con flechas visibles
nx.draw_networkx_edges(
    G, pos,
    arrows=True,
    arrowstyle='->',
    arrowsize=20,
    min_source_margin=20,
    min_target_margin=20
)

# Etiquetas de peso
edge_labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=11)

plt.title("Grafo dirigido ponderado")
plt.axis("off")
plt.show()
```

#### Ponderados en el nodo

```python
G = nx.DiGraph()

G.add_node('Dublin', weight=3)
G.add_node('Paris', weight=5)
G.add_node('Milan', weight=2)
G.add_node('Rome', weight=4)

E = [('Milan','Dublin'), ('Paris','Milan'), ('Paris','Dublin'), ('Milan','Rome')]
G.add_edges_from(E)
```

```python
labels = {n: f"{n}\n({G.nodes[n]['weight']})" for n in G.nodes}

nx.draw(G, pos, 
        with_labels=True, 
        labels=labels,
        node_size=800, 
        node_color='lightgreen', 
        arrows=True)
plt.show()
```

```python
node_sizes = [G.nodes[n]['weight'] * 500 for n in G.nodes]

nx.draw(G, pos, 
        with_labels=True,
        node_size=node_sizes, 
        node_color='lightgreen', 
        arrows=True)
plt.show()
```

#### 2.4 Grafo multipartito

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.DiGraph()

# --- Conjuntos de nodos ---
ciudades = ['Dublin', 'Paris', 'Milan', 'Rome']
autopistas = ['A1', 'A2']
areas_descanso = ['RS1', 'RS2', 'RS3']

# Añadimos nodos con atributo 'part' para indicar el conjunto
G.add_nodes_from(ciudades, part=0)
G.add_nodes_from(autopistas, part=1)
G.add_nodes_from(areas_descanso, part=2)

# --- Aristas entre conjuntos ---
E = [
    ('Paris', 'A1'),
    ('Milan', 'A1'),
    ('Rome', 'A2'),
    ('A1', 'RS1'),
    ('A1', 'RS2'),
    ('A2', 'RS3'),
]

G.add_edges_from(E)
```

```python

# --- Visualización ---
#pos = nx.spring_layout(G, seed=42)
pos = nx.shell_layout(G)
# Colores según el conjunto
color_map = []
for n in G.nodes():
    if G.nodes[n]['part'] == 0:
        color_map.append('skyblue')      # Ciudades
    elif G.nodes[n]['part'] == 1:
        color_map.append('lightgreen')   # Autopistas
    else:
        color_map.append('salmon')       # Áreas de descanso

nx.draw(
    G, pos,
    with_labels=True,
    node_color=color_map,
    node_size=900,
    arrows=True
)

plt.show()
```

In NetworkX, there are several layout algorithms that can be used to position the nodes of a graph in a visualization, including:

- nx.spring_layout(): This algorithm uses a force-directed approach to layout the nodes. Nodes that are connected by edges are attracted to each other, while nodes that are not connected are repelled. The algorithm tries to minimize the energy of the system by adjusting the position of the nodes.  
- nx.circular_layout(): This algorithm positions the nodes evenly around a circle.  
- nx.spectral_layout(): This algorithm uses the eigenvectors of the graph's adjacency matrix to position the nodes. The eigenvectors are used to project the nodes into a lower-dimensional space, and the positions of the nodes are then determined by optimizing a cost function.  
- nx.random_layout(): This algorithm positions the nodes randomly in a given bounding box.  
- nx.shell_layout(): This algorithm positions the nodes in concentric circles or shells, with nodes in the same shell having the same distance to the center.  
- nx.kamada_kawai_layout(): This algorithm uses an iterative optimization approach to layout the nodes. The algorithm tries to minimize the stress of the system by adjusting the position of the nodes.  
- nx.fruchterman_reingold_layout(): This algorithm is a variation of the nx.spring_layout() algorithm, and also uses a force-directed approach to layout the nodes.  

Each layout algorithm has its strengths and weaknesses, and the best one to use depends on the characteristics of the graph and the goals of the visualization. NetworkX allows you to easily apply these layout algorithms to your graphs and create visualizations that help you understand and communicate the structure of the network.
