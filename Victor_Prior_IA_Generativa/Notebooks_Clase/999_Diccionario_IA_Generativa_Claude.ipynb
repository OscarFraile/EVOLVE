### add_messages (Reducer)
Reducer de LangGraph que se aplica al campo de mensajes del estado. En lugar de reemplazar la lista completa cuando un nodo devuelve mensajes nuevos, los **acumula**: añade los nuevos al final de la lista existente.  
**Analogía:** Una bandeja de entrada de email — cada mensaje nuevo se añade, no borra los anteriores.
```python
from langgraph.graph.message import add_messages
from typing import Annotated

class EstadoChat(TypedDict):
    # Sin add_messages: nuevos mensajes REEMPLAZAN la lista
    # Con add_messages: nuevos mensajes se AÑADEN a la lista
    mensajes: Annotated[list[BaseMessage], add_messages]
```

### Embedding
Representación vectorial densa de texto en un espacio de alta dimensión donde palabras similares están cercanas. Es el mecanismo que permite a RAG encontrar documentos relevantes aunque el usuario no use las palabras exactas.

**Ejemplo intuitivo:**
```
"El contrato dura 1 año"    → [0.23, -0.41, 0.87, ...]
"La duración es 12 meses"   → [0.21, -0.39, 0.85, ...] (vectores muy similares)
"El cielo es azul"          → [-0.54, 0.73, -0.31, ...] (vector muy diferente)
```

Parámetro de inferencia también llamado "nucleus sampling". Define el porcentaje acumulado de probabilidad de los tokens que el modelo puede elegir. Complementa a `temperature` para controlar la diversidad de las respuestas.  
**Cómo funciona:** Con `top_p=0.9`, el modelo solo considera los tokens que en conjunto cubren el 90% de la probabilidad, descartando las opciones más improbables.  
**Cuándo ajustarlo:** Por lo general basta con ajustar `temperature`. `top_p` es un segundo nivel de control para casos avanzados.

## Z

### Zero-Shot Prompting ⭐ *nuevo*
Técnica donde el modelo resuelve una tarea **sin ejemplos previos**, solo con instrucciones directas. Es el punto de partida más simple: si funciona, es la solución más eficiente. Si no, se escala a few-shot o chain-of-thought.  
**Cuándo funciona bien:** Tareas directas como clasificar, traducir o resumir con instrucciones claras.  
**Cuándo falla:** Cuando se necesita un formato de salida muy específico o cuando el modelo no sabe "qué tan literal" ser.
```python
prompt = """
Clasifica el sentimiento como POSITIVO, NEGATIVO o NEUTRO.
Responde únicamente con una de esas tres palabras.

Texto: "El servicio fue lento pero la comida estaba deliciosa."
"""
```
**Contrasta con:** Few-Shot (con ejemplos), Chain-of-Thought (con razonamiento).  
**Aparece en:** `[B2]`
