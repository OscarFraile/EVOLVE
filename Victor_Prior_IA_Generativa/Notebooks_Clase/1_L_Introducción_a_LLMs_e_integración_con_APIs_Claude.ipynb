# PARTE 1: FUNDAMENTOS DE IA GENERATIVA Y LLMs

## 1.1 ¿Qué es la IA Generativa?

### Definición

La **IA Generativa** es una rama de la inteligencia artificial capaz de **crear contenido original**: texto, código, imágenes, música y más.

### Características clave

- Utiliza modelos avanzados como **Transformers** y **LLMs** (Large Language Models)
- A diferencia de la IA clásica (que solo analiza datos), **genera resultados creativos**
- Es un subconjunto del **Machine Learning**, que se basa en métodos de **Deep Learning**

### ¿Por qué aprender esto ahora?

1. Los LLMs están transformando múltiples industrias
2. Su adopción es rápida pero la comprensión sigue siendo superficial
3. Este curso enseña desde dentro cómo funcionan y cómo aplicarlos con criterio

### Enfoque del curso: Aprender Haciendo

- 100% práctico, sin relleno
- Lo justo y necesario para aplicar IA generativa en el mundo real

## 1.2 Modelos Discriminativos vs Generativos

| Tipo | Qué hace | Ejemplo |
|------|----------|---------|
| **Discriminativos** | Asocian un dato proporcionado a una respuesta | Clasificar si un email es spam o no |
| **Generativos** | Generan nuevos datos basados en una distribución probabilística de respuestas | Crear un email nuevo, generar una imagen, escribir código |

**Analogía simple:**
- Discriminativo: Como un portero que decide si dejarte entrar o no
- Generativo: Como un artista que crea algo nuevo desde cero

## 1.3 Evolución Histórica de los Modelos de Lenguaje

### Línea del tiempo

```
2013: Word2Vec (representaciones vectoriales de palabras)
   ↓
2015: Seq2Seq con atención (Google Translate)
   ↓
2017: Transformer - "Attention is All You Need" (REVOLUCIÓN)
   ↓
2018-2020: BERT, GPT-2, T5
   ↓
2020-2023: GPT-3, Codex, PaLM, LLaMA
   ↓
2023+: GPT-4, Claude, Gemini, Mistral
```

### Limitaciones de los modelos anteriores (RNN, LSTM, GRU)

| Problema | Descripción |
|----------|-------------|
| **Olvido a largo plazo** | Las RNN olvidan información de palabras lejanas |
| **Lentitud** | Procesan datos secuencialmente, uno tras otro |
| **Difíciles de paralelizar** | No aprovechan bien las GPUs modernas |
| **LSTM/GRU** | Mejoran la memoria pero siguen siendo lentos |

### ¿Qué aportan los Transformers?

1. **Procesan entradas en paralelo** (no paso a paso)
2. **Permiten entrenar modelos con miles de millones de parámetros**
3. **Usan el mecanismo de atención** para enfocarse en partes relevantes
4. **Son mucho más eficientes y escalables** que sus predecesores

### Paper clave

**"Attention is All You Need" (2017)** - Artículo revolucionario que propuso una arquitectura sin recurrencias ni convoluciones, basada únicamente en atención. Desde entonces, es la base para BERT, GPT y T5.

# PARTE 2: ARQUITECTURA TRANSFORMER

## 2.1 ¿Qué es un Transformer?

Es una **arquitectura de red neuronal** diseñada para procesar secuencias enteras en paralelo, utilizando mecanismos de atención en lugar de recurrencia. Supone un salto radical en eficiencia y rendimiento.

### Componentes principales

```
┌──────────────────────────────────────┐
│           TRANSFORMER                │
├──────────────────────────────────────┤
│  ENCODER (procesa la entrada)        │
│  └── Embedding + Positional Encoding │
│  └── Multi-Head Attention            │
│  └── Feed-Forward Network            │
│  └── Normalización + Residual        │
├──────────────────────────────────────┤
│  DECODER (genera la salida)          │
│  └── Masked Multi-Head Attention     │
│  └── Encoder-Decoder Attention       │
│  └── Feed-Forward Network            │
│  └── Normalización + Residual        │
└──────────────────────────────────────┘
```

## 2.2 Tokenización y Embeddings

### ¿Qué es la tokenización?

Es el proceso de convertir texto en **tokens** (fragmentos de palabras) que la máquina puede procesar.

**Ejemplo:**
```
Texto: "El gato come pescado"
Tokens: ["El", " gato", " come", " pescado"]
IDs numéricos: [102, 4521, 889, 3341]
```

### ¿Qué son los Embeddings?

Son **representaciones vectoriales densas** de palabras, frases o documentos. Capturan relaciones semánticas y contextuales.

**Analogía:** Si las palabras fueran ciudades, los embeddings serían sus coordenadas GPS en un mapa. Palabras similares están cerca en el espacio vectorial.

**Visualización del proceso:**
```
Texto humano → Tokens → IDs numéricos → Vectores (embeddings)
   "hola"   → [" hola"] → [4821] → [0.23, -0.45, 0.89, ...]
```

## 2.3 Flujo de Datos dentro del Transformer

```
Paso 1: Input → Se convierte en embeddings (vectores)
   ↓
Paso 2: Se suma la codificación posicional (dónde está cada palabra)
   ↓
Paso 3: Se aplica atención múltiple (multi-head attention)
   ↓
Paso 4: Se procesa con una red feed-forward
   ↓
Paso 5: Se normaliza y pasa a la siguiente capa
   ↓
Paso 6: En el decoder, se usa "self-attention" y "encoder-decoder attention"
   ↓
Output: Texto generado
```

### ¿Por qué funciona tan bien esta arquitectura?

| Característica | Beneficio |
|----------------|-----------|
| **Paralelización total** | Aprovecha mejor las GPUs |
| **Atención directa** | Accede a cualquier parte del input sin importar la distancia |
| **Escalabilidad** | Permite modelos con miles de millones de parámetros |
| **Modularidad** | Se adapta a tareas diversas (traducción, código, imágenes) |

### Proceso de los TRANSFORMERS explicado NO-CODE

Es un proceso que tiene dos partes y en cada parte hay equipos de "personas". El equipo que recibe informacion (encoders) y el de generar informacion (decoders)

Los encoders lo que hacen es coger un texto, lo rompen en trozos mas pequeños (tokenizacion), lo convierten en numeros identificativos (embedding), toman nota del orden de las palabras para que el texto no pierda sentido (positional encoding) y esto lo hacen con todas las palabras al mismo tiempo donde puede haber diferentes significados para cada palabra (Multi-Head Attention), a continuacion cada componente del equipo enconders saca sus conclusiones sobre el significado de lo que ha leido (Feed-Forward) y compara sus conclusiones con sus notas originales guardando una copia de seguridad evitando distorsiones de la informacion cuando el proceso es muy largo (Conexiones Residuales + Normalización)

Los decoders lo que hacen es generar una respuesta y para ello miran lo que dijeron los otros decoders (Masked Attention), preguntan a los encoders que palabra es la mas relevante para poner a continuación (Encoder-Decoder Attention), cada componente del equipo de decoders saca sus conclusiones sobre cual debería ser la siguiente palabra en la respuesta (Feed-Forward) y compara sus conclusiones con sus notas originales guardando una copia de seguridad evitando distorsiones de la informacion cuando el proceso es muy largo (Conexiones Residuales + Normalización)

RESULTADO

- Entrada: "Qué come el gato?"
   ↓
- [El Transformer procesa todo en paralelo]
   ↓
- Salida: "El gato come pescado"

# PARTE 3: MECANISMO DE ATENCIÓN

## 3.1 ¿Qué es la Atención en IA?

Es la **capacidad del modelo para decidir a qué partes de la entrada debe "prestar atención"** para entender o generar una respuesta. Permite enfocarse en la información más relevante.

**Analogía:** Como cuando lees un texto y tus ojos saltan a las palabras clave, ignorando el relleno.

## 3.2 El Núcleo del Transformer: Atención Escalada (Scaled Dot-Product Attention)

### Tres vectores fundamentales

| Vector | Símbolo | Función | Analogía |
|--------|---------|---------|----------|
| **Query** | Q | Lo que busco | Tu pregunta |
| **Key** | K | Identificador del contenido | Etiquetas de archivos |
| **Value** | V | El contenido real | El contenido de los archivos |

### Proceso paso a paso

```
1. Calcular similitud: Q × K^T (producto punto)
   ↓
2. Escalar: dividir por √d_k (dimension de K)
   ↓
3. Normalizar: aplicar Softmax (convierte en probabilidades)
   ↓
4. Ponderar: multiplicar por V
   ↓
5. Resultado: combinación ponderada de los valores
```

**Fórmula:** Attention(Q, K, V) = softmax(QK^T / √d_k) × V

## 3.3 Atención Multi-Cabeza (Multi-Head Attention)

El modelo aplica **múltiples mecanismos de atención en paralelo**. Cada "cabeza" (head) se enfoca en una relación distinta dentro del texto.

**Ejemplo práctico:**
- Cabeza 1: Se enfoca en relaciones gramaticales
- Cabeza 2: Se enfoca en relaciones semánticas
- Cabeza 3: Se enfoca en referencias de entidades
- Cabeza 4: Se enfoca en contexto temporal

```
Input → [Head 1] ──┐
      → [Head 2] ──┤
      → [Head 3] ──┼→ Concatenar → Proyectar → Output
      → [Head 4] ──┤
      → [Head 5] ──┘
      → [Head 6] ──┘
      → [Head 7] ──┘
      → [Head 8] ──┘
```

Esto permite captar **significados complejos y relaciones cruzadas**.

## 3.4 Atención: El Superpoder de los Transformers

### Ventajas clave

1. **Captura dependencias a largo plazo sin esfuerzo**
   - En una frase larga, puede relacionar el sujeto con el verbo al final

2. **No necesita leer en orden: puede mirar toda la secuencia**
   - A diferencia de las RNN que leían palabra por palabra

3. **Permite que el modelo entienda contexto, ambigüedad y relaciones semánticas**
   - Resuelve qué "banco" significa en cada contexto (financiero o asiento)

# PARTE 4: LARGE LANGUAGE MODELS (LLMs)

## 4.1 ¿Qué son los LLMs?

Los **LLMs (Large Language Models)** son modelos entrenados con **enormes cantidades de texto** para predecir y generar lenguaje humano. Son la base de sistemas como ChatGPT, Claude, Gemini o LLaMA.

### ¿Qué pueden hacer los LLMs?

| Capacidad | Ejemplo |
|-----------|---------|
| Generar texto coherente | Escribir un ensayo completo |
| Responder preguntas complejas | Explicar la teoría de la relatividad |
| Traducir entre idiomas | Inglés → Japonés |
| Resumir documentos extensos | Convertir 50 páginas en 1 párrafo |
| Escribir código, fórmulas, emails | Generar una función Python |
| Simular conversaciones | Actuar como un experto en medicina |

### Capacidades cognitivas emergentes

1. **Razonamiento lógico** - Resolver problemas paso a paso
2. **Comprensión de instrucciones** - Seguir prompts detallados
3. **Memoria contextual** - Mantener coherencia en diálogos largos
4. **Adaptabilidad** - Cambiar de tono, formato o estilo según el caso

## 4.2 Aplicaciones en el Mundo Real

| Sector | Aplicación |
|--------|------------|
| **Atención al cliente** | Chatbots, resúmenes de tickets, clasificación de consultas |
| **Legal y compliance** | Análisis de contratos, redacción asistida, verificación de riesgos |
| **Marketing y contenido** | Redacción de textos, generación de copies y campañas |
| **Educación** | Tutores virtuales, generación de tests, explicaciones simplificadas |
| **Programación** | Generación de código, refactorización, documentación automática |

## 4.3 Limitaciones Actuales de los LLMs

| Limitación | Descripción | Ejemplo |
|------------|-------------|---------|
| **Alucinaciones** | Inventan información que suena real pero no lo es | Afirmar que Gabriel García Márquez ganó el Nobel en 1971 (fue 1982) |
| **Sesgos** | Reflejan prejuicios del corpus de entrenamiento | Generar estereotipos de género |
| **Falta de sentido común** | No entienden, solo predicen | No saben que 3+3=6, predicen el token más probable |
| **Actualización limitada** | No siempre conocen eventos recientes | Desconocer noticias de la última semana |

### ¿Qué son las alucinaciones?

Son **errores donde la IA inventa datos que suenan reales pero no lo son**.

**¿Por qué ocurren?**
- La IA predice la palabra más probable, no la verdad
- No piensa, no sabe, no verifica
- Genera frases que suenan correctas pero pueden ser falsas

**Ejemplo visual del proceso:**
```
Usuario: ¿Qué mamífero pone huevos?
   ↓
Modelo: Predice token por token
   ↓
Output 1: "El ornitorrinco es uno de los pocos mamíferos que pone huevos." ✓ CORRECTO
Output 2: "El armadillo es el único mamífero que pone huevos." ✗ ALUCINACIÓN
```

Mismo input, outputs diferentes porque el modelo **predice**, no **sabe**.

## 4.4 Estrategias para Prevenir Errores

### 1. Validación de hechos
- Siempre contrastar datos, nombres, fechas o cifras con fuentes confiables
- Verificar fechas con fuentes oficiales
- Sospechar de cifras extremas o redondas

### 2. Revisión manual
- Solo una persona puede evaluar el contexto, tono e intención correcta
- Leer y editar el contenido final es indispensable

### 3. Uso como borrador, no como fuente definitiva
- La IA es excelente para generar primeros borradores e ideas iniciales
- No debe ser usada como única fuente de información
- Requiere revisión y ajustes humanos

**Ejemplo práctico de validación:**
```
Texto generado por IA:
"El Zoológico de Berlín es uno de los más antiguos del mundo, fundado en 1745.
Alberga más de 30.000 especies... ha sido visitado por más de 100 millones de personas."

Validación:
- ❌ Fundado en 1844 (no 1745)
- ❌ 1.200 especies (no 30.000)
- ⚠️ Buena base, pero debe ajustarse antes de usarse
```

## 4.5 ¿Cómo Generan Texto los LLMs?

Los modelos **no escriben de forma consciente**: predicen el siguiente token basándose en los anteriores.

### Proceso de generación paso a paso

```
Paso 1: Entrada (texto del usuario)
   ↓
Paso 2: Tokenización (dividir en fragmentos)
   ↓
Paso 3: Transformer (atención al contexto)
   ↓
Paso 4: Predicción (generar palabra por palabra)
   ↓
Paso 5: Output (texto coherente)
```

### Predicción token a token

1. El modelo no genera frases completas, sino **un token a la vez**
2. En cada paso, calcula la probabilidad de todos los posibles siguientes tokens
3. Selecciona uno y lo añade al contexto
4. Repite el proceso hasta alcanzar un límite o un token de cierre

**Visualización:**
```
Input: "El gato"
   ↓
Predicciones: come(45%), duerme(30%), salta(15%), maulla(10%)...
   ↓
Selecciona: "come"
   ↓
Nuevo input: "El gato come"
   ↓
Predicciones: pescado(60%), ratones(25%), croquetas(15%)...
   ↓
Continúa hasta completar la respuesta
```

## 4.6 Estrategias de Sampling: Cómo se Elige el Siguiente Token

| Estrategia | Descripción | Cuándo usarla |
|------------|-------------|---------------|
| **Greedy Decoding** | Siempre elige el token más probable | Respuestas directas y factibles |
| **Top-k Sampling** | Elige aleatoriamente entre los k tokens más probables | Balance entre calidad y diversidad |
| **Top-p (Nucleus Sampling)** | Elige de un subconjunto acumulado hasta alcanzar p% de probabilidad | Control fino de diversidad |
| **Temperatura** | Controla la aleatoriedad (baja = más preciso, alta = más creativo) | Ajustar creatividad vs precisión |

### Comportamiento según temperatura

- **Temperatura 0.1**: Texto directo, factual, repetitivo
- **Temperatura 0.9**: Más creativo, impredecible
- **Top-p 0.8**: Tiende a generar resultados más diversos que greedy decoding

### Tipos de modelos según dirección

| Tipo | Dirección | Uso típico | Ejemplo |
|------|-----------|------------|---------|
| **Causal (Autoregresivos)** | Solo hacia adelante | Generación de texto | GPT |
| **Bidireccional** | Contexto completo (antes y después) | Clasificación y comprensión | BERT |

# PARTE 6: INTEGRACIÓN CON APIs

## 6.1 ¿Qué es una Clave de API?

Una **clave de API** es un identificador único que permite autenticar a un usuario o aplicación cuando realiza peticiones a un servicio.

### Funciones de la clave API

1. **Controlar el acceso** a los modelos de IA
2. **Registrar el uso** para monitoreo y facturación
3. **Establecer límites** y políticas de seguridad

## 6.2 Cómo Obtener una Clave de API

### OpenAI
1. Crear cuenta en https://platform.openai.com
2. Ir a API keys → Create new secret key
3. Guardarla (solo se muestra una vez)

### HuggingFace
1. Crear cuenta en https://huggingface.co
2. Ir a Settings → Access Tokens → New Token

⚠️ **Importante**: Nunca compartas ni subas tu clave a repositorios públicos

## 6.3 Formas de Autenticarse en una API

### Método 1: Bearer Token (más común y seguro)

La clave se incluye en el encabezado de la petición HTTP:

```
Authorization: Bearer TU_API_KEY
```

### Método 2: Parámetro en URL (menos seguro)

```
https://api.ejemplo.com/endpoint?api_key=TU_API_KEY
```

### Método 3: Variables de entorno (recomendado)

Guardar la clave en un archivo `.env` y cargarla desde el código:

```python
from dotenv import load_dotenv
import os

load_dotenv()
API_KEY = os.getenv("API_KEY")
```

⚠️ **Siempre usar HTTPS** para proteger tu clave en tránsito

## 6.4 Buenas Prácticas en el Uso de Claves API

| Práctica | Descripción |
|----------|-------------|
| **No publiques tu clave** | Ni en GitHub ni en notebooks compartidos |
| **Usa variables de entorno** | Guardarlas en `.env`, nunca en el código |
| **Rota las claves periódicamente** | Generar nuevas claves cada cierto tiempo |
| **Revoca si se filtra** | Desactivar inmediatamente si hay sospecha de filtración |
| **Revisa los logs de uso** | Monitorear desde el panel del proveedor |

````python
# Ejemplo de configuración segura de API key
from dotenv import load_dotenv
import os

# Carga las variables de entorno
load_dotenv()

# Obtiene la API key de forma segura
API_KEY = os.getenv("OPENAI_API_KEY")

# Verifica que se cargó correctamente (sin mostrarla)
if API_KEY:
    print("✅ API Key cargada correctamente")
    print(f"   Longitud: {len(API_KEY)} caracteres")
    print(f"   Primeros 10: {API_KEY[:10]}...")
else:
    print("❌ No se encontró la API Key")
````

````python
# Ejemplo de llamada a API con manejo de errores
import requests

def llamar_api_openai(prompt, api_key):
    """
    Realiza una llamada a la API de OpenAI de forma segura.
    
    Args:
        prompt: El texto de entrada para el modelo
        api_key: La clave de API (cargada desde variables de entorno)
    
    Returns:
        La respuesta del modelo o un mensaje de error
    """
    url = "https://api.openai.com/v1/chat/completions"
    
    headers = {
        "Authorization": f"Bearer {api_key}",
        "Content-Type": "application/json"
    }
    
    data = {
        "model": "gpt-3.5-turbo",
        "messages": [{"role": "user", "content": prompt}],
        "temperature": 0.7
    }
    
    try:
        response = requests.post(url, headers=headers, json=data)
        response.raise_for_status()
        return response.json()["choices"][0]["message"]["content"]
    except requests.exceptions.RequestException as e:
        return f"Error en la llamada: {e}"

# Ejemplo de uso (comentado para no consumir tokens)
# resultado = llamar_api_openai("Explica qué es un LLM", API_KEY)
# print(resultado)
````

## Evolución en una Línea

```
RNN (secuencial, lento) → LSTM/GRU (mejor memoria) → Transformer (paralelo, atención) → LLMs (GPT, Claude, Gemini)
```

## Arquitectura Transformer Resumida

```
Input → Embeddings + Positional Encoding → Multi-Head Attention → Feed-Forward → Normalización → Output
```

## Precauciones Importantes

1. ⚠️ Los LLMs **predicen**, no **saben** ni **entienden**
2. ⚠️ Siempre **validar** información factual (fechas, nombres, cifras)
3. ⚠️ Usar como **borrador**, no como fuente definitiva
4. ⚠️ **Nunca** exponer API keys en código público
5. ⚠️ Usar **variables de entorno** para proteger credenciales

# DICCIONARIO DE TÉRMINOS - INTRODUCCIÓN A LLMs

Glosario de conceptos clave del módulo de Introducción a LLMs e integración con APIs.

---

## A

| Término | Definición |
|---------|------------|
| **Alucinación** | Error donde la IA inventa información que suena real pero no lo es. Ocurre porque el modelo predice el token más probable, no la verdad. |
| **API (Application Programming Interface)** | Interfaz que permite que diferentes aplicaciones se comuniquen entre sí. En IA, permite integrar modelos en aplicaciones. |
| **API Key** | Clave de identificación única que autentica a un usuario o aplicación al realizar peticiones a un servicio de API. |
| **Atención (Attention)** | Mecanismo que permite al modelo decidir a qué partes de la entrada debe "prestar atención" para entender o generar una respuesta. |
| **Atención Multi-Cabeza (Multi-Head Attention)** | Aplicación de múltiples mecanismos de atención en paralelo. Cada cabeza se enfoca en relaciones distintas dentro del texto. |

---

## B

| Término | Definición |
|---------|------------|
| **BERT** | Modelo bidireccional de Google (2018) que revolucionó las tareas de comprensión de lenguaje. |
| **Bidireccional** | Modelo que puede ver todo el contexto (antes y después). Ejemplo: BERT. Usado para comprensión y clasificación. |
| **Bearer Token** | Método de autenticación donde la clave API se incluye en el encabezado HTTP: "Authorization: Bearer TU_API_KEY". |

---

## C

| Término | Definición |
|---------|------------|
| **Claude** | Modelo de Anthropic enfocado en seguridad y alineación con valores humanos. |
| **Codificación Posicional** | Información añadida a los embeddings para indicar la posición de cada palabra en la secuencia, ya que los Transformers no procesan en orden. |
| **Conexiones Residuales** | Conexiones "skip" que suman la entrada de una capa a su salida, permitiendo que el gradiente fluya mejor durante el entrenamiento. |
| **Corpus** | Conjunto de textos utilizados para entrenar un modelo de lenguaje. |
| **Causal (Autoregresivo)** | Modelo que solo puede ver tokens anteriores para predecir el siguiente. Ejemplo: GPT. Usado para generación de texto. |

---

## D

| Término | Definición |
|---------|------------|
| **Decoder** | Componente del Transformer que genera la salida (texto destino) basándose en las representaciones del encoder. Usado en generación. |
| **Discriminativo** | Modelo que asocia un dato de entrada a una etiqueta o clase. Ejemplo: clasificar si un email es spam o no. |

---

## E

| Término | Definición |
|---------|------------|
| **Embedding** | Representación vectorial densa de una palabra, frase o documento. Captura relaciones semánticas en un espacio numérico de alta dimensión. |
| **Encoder** | Componente del Transformer que procesa la entrada (texto de origen) y crea representaciones internas. Usado en tareas de comprensión. |

---

## F

| Término | Definición |
|---------|------------|
| **Fine-tuning** | Proceso de ajustar un modelo pre-entrenado con datos específicos de una tarea o dominio particular. |
| **Feed-Forward** | Capas neuronales densas que procesan la información después de la atención, añadiendo no-linealidad al modelo. |

---

## G

| Término | Definición |
|---------|------------|
| **Gemini** | Modelo de Google DeepMind, multimodal (texto + imagen). |
| **Generativo** | Modelo que genera nuevos datos basados en una distribución probabilística. Ejemplo: crear un email, generar una imagen, escribir código. |
| **GPT (Generative Pre-trained Transformer)** | Familia de modelos de OpenAI. GPT-4 es uno de los modelos más avanzados actualmente. |
| **Greedy Decoding** | Estrategia que siempre elige el token más probable. Produce respuestas directas pero puede ser repetitivo. |
| **GRU** | Variante de RNN con mejores mecanismos de memoria, pero aún secuencial y lenta comparada con Transformers. |

---

## H

| Término | Definición |
|---------|------------|
| **HTTPS** | Protocolo seguro de transferencia web que encripta la comunicación, esencial para proteger API keys en tránsito. |

---

## I

| Término | Definición |
|---------|------------|
| **IA Generativa** | Rama de la IA capaz de crear contenido original: texto, código, imágenes, música. A diferencia de la IA clásica que solo analiza, esta genera resultados creativos. |
| **Inferencia** | Proceso de usar un modelo entrenado para hacer predicciones o generar respuestas ante nuevas entradas. |

---

## K

| Término | Definición |
|---------|------------|
| **Key (K)** | Vector que actúa como identificador del contenido. Permite encontrar qué partes del input son relevantes para la query. |

---

## L

| Término | Definición |
|---------|------------|
| **LLM (Large Language Model)** | Modelo de lenguaje grande entrenado con enormes cantidades de texto para predecir y generar lenguaje humano. Ejemplos: GPT-4, Claude, Gemini. |
| **LLaMA** | Modelo open source de Meta que puede alojarse localmente. |
| **LSTM** | Variante de RNN con mejores mecanismos de memoria, pero aún secuencial y lenta comparada con Transformers. |

---

## M

| Término | Definición |
|---------|------------|
| **Mistral** | Modelo open source ligero, rápido y eficiente. |
| **Modelo** | En el contexto de ML, arquitectura matemática que aprende patrones de los datos. |

---

## N

| Término | Definición |
|---------|------------|
| **Normalización (LayerNorm)** | Técnica que estabiliza el entrenamiento normalizando las activaciones en cada capa del modelo. |

---

## P

| Término | Definición |
|---------|------------|
| **Paper "Attention is All You Need"** | Artículo de 2017 que introdujo la arquitectura Transformer. Punto de inflexión en IA. |
| **Parámetro** | Valor ajustable en el modelo que se aprende durante el entrenamiento. Los LLMs tienen miles de millones de parámetros. |
| **Prompt** | Texto de entrada que se proporciona al modelo para obtener una respuesta. Instrucción o pregunta para el LLM. |
| **Prompt Engineering** | Diseño y optimización de prompts para obtener respuestas más precisas y útiles del modelo. |

---

## Q

| Término | Definición |
|---------|------------|
| **Query (Q)** | Vector que representa "lo que busco" en el mecanismo de atención. Es la pregunta que se hace al modelo. |

---

## R

| Término | Definición |
|---------|------------|
| **Red Feed-Forward** | Capas neuronales densas que procesan la información después de la atención, añadiendo no-linealidad al modelo. |
| **RNN (Red Neuronal Recurrente)** | Arquitectura anterior que procesaba secuencias paso a paso. Tenía problemas de olvido a largo plazo y era lenta. |

---

## S

| Término | Definición |
|---------|------------|
| **Scaled Dot-Product Attention** | Método matemático para calcular la atención: softmax(QK^T / √d_k) × V. Normaliza y pondera la relevancia de cada parte del input. |
| **Seq2Seq** | Arquitectura secuencia a secuencia usada en traducción automática antes de los Transformers. |

---

## T

| Término | Definición |
|---------|------------|
| **Temperatura** | Parámetro que controla la aleatoriedad en la generación. Valores bajos (0.1) = más preciso/predictible. Valores altos (0.9) = más creativo/impredecible. |
| **Token** | Unidad básica de texto que el modelo procesa. Puede ser una palabra completa, parte de una palabra o un carácter. |
| **Tokenización** | Proceso de convertir texto en tokens (fragmentos de palabras) que la máquina puede procesar. Ejemplo: "El gato" → ["El", " gato"] → [102, 4521]. |
| **Top-k Sampling** | Estrategia que elige aleatoriamente entre los k tokens más probables. Balance entre calidad y diversidad. |
| **Top-p (Nucleus Sampling)** | Estrategia que elige de un subconjunto acumulado hasta alcanzar p% de probabilidad total. Control fino de diversidad. |
| **Transfer Learning** | Aprovechar conocimientos aprendidos en una tarea para aplicarlos a otra tarea relacionada. Base del fine-tuning. |
| **Transformer** | Arquitectura de red neuronal que procesa secuencias en paralelo usando mecanismos de atención, en lugar de recurrencia. Revolucionó el procesamiento de lenguaje natural desde 2017. |

---

## V

| Término | Definición |
|---------|------------|
| **Value (V)** | Vector que contiene el contenido real que se extraerá una vez identificada la relevancia mediante Q y K. |
| **Variable de Entorno** | Mecanismo para almacenar configuraciones sensibles (como API keys) fuera del código fuente, típicamente en archivos .env. |

---

## W

| Término | Definición |
|---------|------------|
| **Word2Vec** | Modelo de 2013 que crea representaciones vectoriales de palabras. Pionero en embeddings. |
