```python
# ============================================
# CONFIGURACIÓN COMPLETA: Proteger API Keys
# ============================================

# %% 1. INSTALAR DEPENDENCIA (solo primera vez)
# python-dotenv carga variables desde archivo .env automáticamente
!pip install python-dotenv -q

# %% 2. CREAR ARCHIVO .env CON TU API KEY
# Este archivo guarda secrets LOCALMENTE. NUNCA se sube a GitHub.
env_content = """# Variables de entorno - NO SUBIR A GIT
OPENAI_API_KEY=sk-tu-clave-aqui
"""
with open(".env", "w") as f:
    f.write(env_content)
print("✅ Archivo .env creado")

# %% 3. AÑADIR .env A .gitignore
# .gitignore le dice a Git qué archivos ignorar (no trackear)
gitignore_content = """# Secrets y variables de entorno
.env
.env.local
.env.production
"""
with open(".gitignore", "a") as f:
    f.write(gitignore_content)
print("✅ .env añadido a .gitignore")

# %% 4. CARGAR LA API KEY EN TU CÓDIGO
from dotenv import load_dotenv
import os

load_dotenv()  # Carga variables del archivo .env a environment

api_key = os.getenv("OPENAI_API_KEY")
print(f"✅ API Key cargada: {api_key[:10]}... (oculto por seguridad)")

# %% 5. VERIFICAR QUE GIT LO IGNORA CORRECTAMENTE
import subprocess
result = subprocess.run(["git", "check-ignore", ".env"], capture_output=True, text=True)
if result.returncode == 0:
    print("✅ Git está ignorando .env correctamente")
else:
    print("⚠️  Git NO está ignorando .env. Revisa .gitignore")
```
