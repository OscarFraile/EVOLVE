### Reduce

````python
from functools import reduce
````

Esta función no es una función incorporada, lo que significa que primero debe importar la biblioteca functools para usarla. La función ``
`reduce()` aplica de forma acumulativa la función f() a los elementos del iterable. El acumulador se puede inicializar con un tercer argumento opcional. El resultado es el valor final del acumulador.

```python
reduce(function, iterable[, initial])
```

#### ¿Qué sucedió?
- Empezamos con una lista [2, 4, 7, 3] y pasamos la función add (x, y) a reduce ( ) junto con esta lista, sin un valor inicial

- reduce() llama a add(2, 4), y add() devuelve 6

- reduce() llama a sum(6, 7) (resultado de la llamada anterior a sum() y el siguiente elemento en la lista como parámetros), y sum() devuelve 13

- reduce() llama a sum(13, 3), y sum() devuelve 16

Como no quedan más elementos en la secuencia, reduce() devuelve 16
