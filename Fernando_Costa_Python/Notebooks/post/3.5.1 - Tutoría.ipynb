```python
class Staff ():

    # 1. ATRIBUTOS
    def __init__ (self, nombre, apellido, ciudad, vacaciones=30):
        # 1.1. Generales
        self.nombre = nombre
        self.apellido = apellido.lower()
        self.ciudad = ciudad

        # 2. Construidos
        self.email = f"{nombre.lower()}.{self.apellido}@empresa.com" # aquí hace falta self según la lógica 

        # 3. Para modificar luego
        self.vacaciones = vacaciones
    
    # 2. MÉTODOS
    def saludar (self):
        return f"Hola, soy {self.nombre} {self.apellido} y vivo en {self.ciudad}"
    
    def cogervacaciones (self, dias=1):
        if dias < 0:
            raise ValueError("No puedes des-cogerte días")

        if self.vacaciones >= dias:
            self.vacaciones -= dias
            return self.vacaciones
        else:
            return f"No puedes cogerte {dias} de vacaciones porque te quedan {self.vacaciones}"
```

```python
class Marketing (Staff):
    
    def publicar ():
        return f"He subido algo a instagram"
```

```python
class Sales (Staff):
    def __init__(self, nombre, apellido, ciudad):
        super().__init__(nombre, apellido, ciudad)
        self.total_ventas = 13
    
    def vender (self, ventas_hechas):
        self.total_ventas += ventas_hechas
        return f"He vendido un total de {self.total_ventas} porque hoy he hecho {ventas_hechas}"
```

```python
class Intern (Sales, Marketing):
    
    def __init__ (self, nombre, apellido, ciudad,  vacaciones=3):
        super().__init__(nombre, apellido, ciudad)
        self.vacaciones = vacaciones
```

```python
alvaro = Staff("Álvaro", "González", "Madrid", vacaciones = 10)
carlos = Marketing("Carlos", "Martínez", "Álava", vacaciones=10)
laura = Sales ("Laura", "González", "Valencia")
rosa = Intern("Rosa", "Serrano", "Badajoz")
```
