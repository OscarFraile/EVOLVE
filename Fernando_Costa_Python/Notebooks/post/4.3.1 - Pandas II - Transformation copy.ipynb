### df.loc y df.iloc

#### Basado en la etiqueta: `loc`

El m√©todo loc se puede utilizar de dos maneras diferentes: seleccionar filas o columnas seg√∫n una etiqueta, o seleccionar filas o columnas seg√∫n una condici√≥n.

#### Basado en el √≠ndice: `iloc`

El m√©todo iloc se utiliza en DataFrames para seleccionar elementos en funci√≥n de su ubicaci√≥n. Su sintaxis es data.iloc[<filas>, <columnas>], donde <filas> y <columnas> son la posici√≥n de las filas y columnas que se quieren seleccionar en el orden en el que aparecen en el objeto. Una notaci√≥n familiar para los usuarios de Matlab. En un DataFrame, cada una de las filas tiene un n√∫mero que va desde 0 hasta el n√∫mero total de filas menos uno. Siendo iloc el m√©todo que permite seleccionar los datos en funci√≥n de estos n√∫meros. Lo mismo aplica para las columnas.

En iloc hay dos argumentos filas y columnas. En caso de omitir el segundo, se seleccionar√°n todas las columnas de la fila. Por lo que para seleccionar diferentes filas se pueden utilizar.

Sintaxis loc con m√∫ltiples condiciones
```python
df.loc[(df["column"] condition) & (df["column"] condition)]
````
