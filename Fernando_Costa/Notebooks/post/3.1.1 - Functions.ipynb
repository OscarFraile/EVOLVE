````python
def factura (importe_bruto, retencion, *args): #* n煤mero variable de argumento, tan variable que puede ser 0
    resultado = importe_bruto * retencion
    for i in args:
        resultado += i
    return resultado

factura(50, 0.85, 76, 23, 1)
````

### `lambda` como opci贸n de algunas otras funciones integradas

Tambi茅n es posible utilizar estas funciones an贸nimas como argumentos en llamadas a otras funciones. Por ejemplo, en una llamada a sorted() o list.sort() podemos utilizar una funci贸n lambda como argumento del par谩metro clave:

````python
def greater_than (a, b):
    return max([a, b])

def less_than (a, b):
    return min([a, b])

def compare (decision):
    if decision == "mayorque":
        return greater_than
    elif decision == "menorque":
        return less_than
````

````python
compare("mayorque")(3, 30)

# 1. compare("mayorque") -> greater_than
# 2. greater_than(3, 30)
````

````python
def compare_lambda (decision):
    if decision == "mayorque":
        return lambda a, b: a if a > b else b
    elif decision == "menorque":
        return lambda a, b: b if b < a else a
    
compare_lambda("menorque")(3, 30)
````

## Recursividad

Se denomina llamada recursiva (o recursi贸n), a aquellas funciones que en su algoritmo, hacen referencia a s铆 misma.

Las llamadas recursivas suelen ser muy 煤tiles en casos muy concretos, pero debido a su gran factibilidad de caer en iteraciones infinitas, se deben tomar las medidas preventivas adecuadas y solo utilizarlas cuando sea estrictamente necesario y no exista una forma alternativa viable de resolver el problema, evitando as铆 la recursi贸n.

Python soporta llamadas recursivas, permitiendo que una funci贸n se llame a s铆 misma, tal y como lo hace al llamar a otra funci贸n.

````python
def numero ():
    """Esta funci贸n hace cosas"""
    print("Ejecutando funci贸n")
    value = int(input (prompt = "Un n煤mero mayor que 5: "))
    if value > 5:
        return value
    else:
        return numero()
````
