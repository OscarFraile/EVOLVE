### Matriz de ceros
np.zeros se utiliza para crear la matriz donde todos los elementos son 0. Su sintaxis es:
`np.zeros(shape, dtype=float)`

Donde:
- La forma es el tama침o de la matriz y puede ser unidimensional, bidimensional o de m칰ltiples dimensiones.
- El dtype es float64 de manera predeterminada, pero se puede configurar en cualquier tipo de datos en NumPy.

### Matriz de UNOS
De manera similar a los ceros, tambi칠n podr칤amos crear una matriz llena de unos. La sintaxis y los par치metros de np.ones() son id칠nticos a los de np.zeros().

### Matriz DIAGONAL
Devuelve una matriz bidimensional con unos en la diagonal y ceros en el resto. La sintaxis es:
`np.eye(N, M=None, k=0, dtype=<class 'float'>, order='C')`
Donde:
- **N**: INT Especifica el n칰mero de l칤neas en la matriz generada.
- **M**: INT (opcional) El valor predeterminado es None. Especifica el n칰mero de columnas en la matriz que se generar치. Cuando es None, es igual a N.
- **k**: INT (opcional) El valor predeterminado es 0. 0 es la posici칩n principal de la diagonal. Si k es positivo, crea la diagonal superior con el desplazamiento de k; si k es negativo, crea la diagonal inferior con el desplazamiento de k.
- **dtype**: tipo de datos (opcional) El valor predeterminado es float. Especifica el tipo de datos del elemento de la matriz creada.

```python
def picture_info(pic):
    
    print(f"Shape of the picture is: {pic.shape}")

    print(f"1. First dimension: {pic.shape[0]} pixels on Y axis\n\n Example: {pic[0]}\n") 
    print(f"2. Second dimension: {pic.shape[1]} pixels on X axis\n\nExample: {pic[0]}\n") 

    print("--------------------------\n")
    print(f"2. Third dimension (RGB) is: {pic[0][0]}\n") 
    print("--------------------------\n")
    print(f"- Red: {pic[0][0][0]}")
    print(f"- Green: {pic[0][0][1]}") 
    print(f"- Blue: {pic[0][0][2]}")
    
picture_info(photo)
```
