## 7. Write a function that takes a list and returns a list of the unique values.

Forbidden built-in functions: `set()` nor `lst.append`

```python
def unique(arr):
    # iterar por cada n√∫mero

    dict_ = {}
    for i in arr:
        dict_[i] = arr.count(i)

    return list(dict_.keys())


unique([1,2,2,3,4,4,5,5,5,6])
```

```python
def unique(arr):

    str_ = ""
    for i in arr:
        if str(i) not in str_:
            str_ += str(i)
    return [int(i) for i in str_]

unique([1,2,2,3,4,4,5,5,5,6])
```

```python
def unique(arr):

    unique_list = []

    for i in arr:

        if i not in unique_list:

            unique_list += [i]

    return unique_list


unique([1,2,2,3,4,4,5,5,5,6])
```

## 8. Write a function that calculates the standard deviation of a list.

Forbidden built-in functions: `sum()` and no imports of libraries

```python
def st_dev(arr):
    # Desviaci√≥n t√≠pica: 
        # cu√°nto se desv√≠a de la media
        # Si tienes una media de 1.500‚Ç¨ y una desviaci√≥n de 400‚Ç¨
            # "te puedes equivocar 400 arriba o abajo"
            # "los datos est√°n comprendidos en (1.100, 1.900)" -> no
            # "cu√°nto se separan los datos de la media EN CONJUNTO"
            # dispersi√≥n: cu√°nto se acercan los datos a la media
    # DE MEDIA, cu√°nto se separa cada punto de la media
        # 1. calcular la media
        # 2. diferencia: cada punto con la media
        # 3. La media de las diferencias


    sumador = 0
    longitud = 0

    operacion = 0
    desviacion = 0


    # 1. La media
    for i in arr:
        sumador += i
        longitud +=1

        media =sumador/longitud

    # 2. 
    for k in arr:

        operacion += ((k-media)**2)/longitud


        desviacion = operacion ** 0.5

    return desviacion

arr = [4, 5, 6] # mean: 5, stdev: 0.81 # menos disperso
arr = [1, 5, 9] # mean: 5, stdev: 3.2 # m√°s disperso
print(sum(arr)/len(arr))
st_dev(arr)
```
