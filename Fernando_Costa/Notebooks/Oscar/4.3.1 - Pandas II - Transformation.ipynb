#### Basado en la etiqueta: `loc`

#### Basado en el índice: `iloc`

El método iloc se utiliza en DataFrames para seleccionar elementos en función de su ubicación. Su sintaxis es data.iloc[<filas>, <columnas>], donde <filas> y <columnas> son la posición de las filas y columnas que se quieren seleccionar en el orden en el que aparecen en el objeto. Una notación familiar para los usuarios de Matlab. En un DataFrame, cada una de las filas tiene un número que va desde 0 hasta el número total de filas menos uno. Siendo iloc el método que permite seleccionar los datos en función de estos números. Lo mismo aplica para las columnas.

En iloc hay dos argumentos filas y columnas. En caso de omitir el segundo, se seleccionarán todas las columnas de la fila. Por lo que para seleccionar diferentes filas se pueden utilizar.

Sintaxis loc con múltiples condiciones
```python
df.loc[(df["column"] condition) & (df["column"] condition)]
````

7. **DEFINE** (ENCAPSULA) en una FUNCIÓN
- Nombre de la función: profit_per_customer
- Argumentos de la función: name, string. el nombre del cliente por el que se filtrará
- para hacer esto, deberá cambiar la línea donde filtra para cambiar el literal "Roy Französisch" por name
- cámbielo también en la impresión formateada
- La función debe realizar los pasos: 1-8
- incluir la línea: `df_imported = pd.read_csv("customer_example.csv")`
- Retorno de la función: debe devolver `df_imported`

`````python
def get_names(df):
    # 1º. Cargar el dataframe
    df = pd.read_csv(r'../../../../evolve-data-python\datasets\superstore.csv')

    # 2º. Elimina los guiones y todo minúscula
    df.columns = df.columns.str.replace(' ', '_').str.lower()

    # 3º. Extrae una lista con los nombres que existen y ordenalos alfabéticamente
    df.customer_name = df.customer_name.str.lower()
    names = df.customer_name.unique()
    names.sort()

    return names

nombres = get_names(df)
print(nombres)
`````

`````python
def profit_per_customer(name_input):
    # 1º. Cargar el dataframe
    df = pd.read_csv(r'../../../../evolve-data-python\datasets\superstore.csv')

    # 2º. Elimina los guiones y todo minúscula
    df.columns = df.columns.str.replace(' ', '_').str.lower()

    # 3º. Elimina columna
    df = df.drop(columns=['row_id'])

    # 4º Estandarizar la columna 'customer_name' y el input
    name = name_input.lower()
    df['customer_name'] = df['customer_name'].str.lower()
    #df.customer_name = df.customer_name.str.lower()

    # 5º Calcular la ganancia total para el cliente
    if name not in df['customer_name'].unique():
        return 'Probablemente el nombre no conste en la tabla o esté mal escrito'
    
    else:
        prof = df.loc[(df.customer_name == name)].profit.sum()
        
    return f'{name_input} obtuvo un resultado total de {prof:.2f}$'


# Input de la barra de arriba
name_input = input('Nombre a buscar: ')

# Resultado de la función
resultado = profit_per_customer(name_input)
print(resultado)

`````
