## 1. Write a function that returns the greater of two numbers

```python
def greater(a,b):
    if a > b:
        return a
    else:
        return b
greater(2,3)
```

## 2. Now write a function that returns the largest element on a list

Forbidden built-in functions: `max()` & `sorted()`

```python
def greatest(arr):
    #Cuenta los caracteres de cada elemento de la lista arr y devuelve el elemento mas largo sin la funcion max()
    longest = ''
    for element in arr:
        if len(element) > len(longest):
            longest = element
    return longest

greatest(['Python', 'is', 'awesome', 'JavaScript'])
```

## 6. Write a function that returns the factorial of a number.

Forbidden built-in functions: `math.factorial()`

```python
def factorial(n):
    num = n
    while n > 1:
        n -= 1
        num *= n
    return num
factorial(5)
```

## 8. Write a function that calculates the standard deviation of a list.

Forbidden built-in functions: `sum()` and no imports oif libraries

```python

def st_dev(arr):
    # calcular media sin sum()
    total = 0
    for x in arr:
        total += x
    media = total / len(arr)

    # calcular sumatoria de (x - media)^2 sin sum()
    acumulado = 0
    for x in arr:
        diferencia = x - media
        acumulado += diferencia * diferencia

    # ra√≠z cuadrada sin math: usar ** 0.5
    return (acumulado / len(arr)) ** 0.5
arr = [10,12,23,23,16,23,21,16]
print(st_dev(arr)) #4.898979485566356
print(sum(arr)/len(arr)) #18.0

arr = [10,12,45,1,16,58,1,1]
print(st_dev(arr)) #20.310096011589902
print(sum(arr)/len(arr)) #18.0
```

## 9. Write a function to check if a string is a pangram, i.e.: if it contains all the letters of the alphabet at least once. Mind that the strings may contain characters that are not letters.

You can use `str.lower()`

```python
def pangram(string):
    alphabet = 'abcdefghijklmnopqrstuvwxyz'
    string = string.lower()
    for letter in alphabet:
        if letter not in string:
            return False
    return True
pangram("The quick brown fox jumps over the lazy dog")

    
    
```
