## ¬øQu√© es Numpy?

```python
import numpy as np
```

NumPy es una biblioteca para el lenguaje de programaci√≥n Python que brinda soporte para crear matrices y vectores multidimensionales de gran tama√±o, junto con una gran colecci√≥n de funciones matem√°ticas de alto nivel para operar sobre ellos. El precursor de NumPy, Numeric, fue creado originalmente por Jim Hugunin con contribuciones de varios otros desarrolladores. En 2005, Travis Oliphant cre√≥ NumPy incorporando caracter√≠sticas de la competencia Numarray en Numeric, con amplias modificaciones. NumPy es un software de c√≥digo abierto y cuenta con muchos colaboradores.

El n√∫cleo del paquete NumPy es el objeto ndarray. Este encapsula matrices n-dimensionales de tipos de datos homog√©neos, y muchas operaciones se realizan en c√≥digo compilado para el rendimiento. ## Trabajar con datos num√©ricos
"Datos" en el an√°lisis de datos a menudo se refiere a datos num√©ricos, por ejemplo, precios de acciones, cifras de ventas, mediciones de sensores, resultados deportivos, tablas de bases de datos, etc. La biblioteca Numpy proporciona estructuras de datos especializadas, funciones y otras herramientas para el c√°lculo num√©rico en Python. Trabajemos con un ejemplo para ver por qu√© y c√≥mo usar Numpy para trabajar con datos num√©ricos.

Supongamos que queremos usar datos clim√°ticos como temperatura, precipitaciones y humedad para determinar si una regi√≥n es adecuada para el cultivo de manzanas. Un enfoque simple para hacer esto ser√≠a formular la relaci√≥n entre el rendimiento anual de manzanas (toneladas por hect√°rea) y las condiciones clim√°ticas como la temperatura media (en grados Fahrenheit), las precipitaciones (en mil√≠metros) y la humedad relativa media (en porcentaje) como una ecuaci√≥n lineal.

`yield_of_apples = w1 *temperature + w2* rainy + w3 * humidity`

Estamos expresando el rendimiento de manzanas como una suma ponderada de temperatura, precipitaciones y humedad. Esta ecuaci√≥n es una aproximaci√≥n, ya que la relaci√≥n real puede no ser lineal y pueden estar involucrados otros factores. Pero un modelo lineal simple como este suele funcionar bien en la pr√°ctica.

Con base en un an√°lisis estad√≠stico de datos hist√≥ricos, podemos obtener valores razonables para los pesos w1, w2 y w3. A continuaci√≥n, se muestra un ejemplo de valores:

## Ventajas de usar matrices Numpy

Las matrices de Numpy ofrecen las siguientes ventajas sobre las listas de Python para operar con datos num√©ricos:

- Facilidad de uso: puede escribir expresiones matem√°ticas peque√±as, concisas e intuitivas como `(kanto * weights).sum()` en lugar de usar bucles y funciones personalizadas como crop().
- Rendimiento: las operaciones y funciones de Numpy se implementan internamente en C++, lo que las hace mucho m√°s r√°pidas que usar declaraciones y bucles de Python que se interpretan en tiempo de ejecuci√≥n.

### Matriz de ceros
np.zeros se utiliza para crear la matriz donde todos los elementos son 0. Su sintaxis es:
`np.zeros(shape, dtype=float)`

Donde:
- La forma es el tama√±o de la matriz y puede ser unidimensional, bidimensional o de m√∫ltiples dimensiones.
- El dtype es float64 de manera predeterminada, pero se puede configurar en cualquier tipo de datos en NumPy.

### Matriz de UNOS
De manera similar a los ceros, tambi√©n podr√≠amos crear una matriz llena de unos. La sintaxis y los par√°metros de np.ones() son id√©nticos a los de np.zeros().

### Matriz DIAGONAL
Devuelve una matriz bidimensional con unos en la diagonal y ceros en el resto. La sintaxis es:
`np.eye(N, M=None, k=0, dtype=<class 'float'>, order='C')`
Donde:
- **N**: INT Especifica el n√∫mero de l√≠neas en la matriz generada.
- **M**: INT (opcional) El valor predeterminado es None. Especifica el n√∫mero de columnas en la matriz que se generar√°. Cuando es None, es igual a N.
- **k**: INT (opcional) El valor predeterminado es 0. 0 es la posici√≥n principal de la diagonal. Si k es positivo, crea la diagonal superior con el desplazamiento de k; si k es negativo, crea la diagonal inferior con el desplazamiento de k.
- **dtype**: tipo de datos (opcional) El valor predeterminado es float. Especifica el tipo de datos del elemento de la matriz creada.

```python
def picture_info(pic):
    
    print(f"Shape of the picture is: {pic.shape}")

    print(f"1. First dimension: {pic.shape[0]} pixels on Y axis\n\n Example: {pic[0]}\n") 
    print(f"2. Second dimension: {pic.shape[1]} pixels on X axis\n\nExample: {pic[0]}\n") 

    print("--------------------------\n")
    print(f"2. Third dimension (RGB) is: {pic[0][0]}\n") 
    print("--------------------------\n")
    print(f"- Red: {pic[0][0][0]}")
    print(f"- Green: {pic[0][0][1]}") 
    print(f"- Blue: {pic[0][0][2]}")
    
picture_info(photo)
```
