```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Configuración visual para mantener consistencia
plt.style.use('seaborn-v0_8-whitegrid')
plt.rcParams['figure.figsize'] = [10, 6]
plt.rcParams['font.size'] = 12

# Carga de datos de ejemplo (Seaborn Tips)
tips = sns.load_dataset('tips')
```

```python
tips.describe()
```

```python
plt.hist(tips['total_bill'], bins=20,
         color='#333333', edgecolor='white')
plt.title('Distribución de Facturación')
plt.show()
```

```python
sns.boxplot(
    data=tips,
    x='day',
    y='total_bill'
)
```

```python
plt.scatter(tips['total_bill'], tips['tip'])
plt.xlabel('Factura ($)')
plt.ylabel('Propina ($)')
```
